```python
soup2 = """<!DOCTYPE html>
<html lang="en-US" prefix="og: http://ogp.me/ns#" xmlns:addthis="https://www.addthis.com/help/api-spec" xmlns:fb="https://www.facebook.com/2008/fbml">
<head>
<meta charset="utf-8"/>
<meta content="width=device-width" name="viewport"/>
<title>May 2013 - Basketball Intelligence</title>
<link href="http://gmpg.org/xfn/11" rel="profile"/>
<link href="http://basketballintelligence.net/xmlrpc.php" rel="pingback"/>
<!--[if lt IE 9]>
<script src="http://basketballintelligence.net/wp-content/themes/superhero/js/html5.js" type="text/javascript"></script>
<![endif]-->
<script>(function(d, s, id){
				 var js, fjs = d.getElementsByTagName(s)[0];
				 if (d.getElementById(id)) {return;}
				 js = d.createElement(s); js.id = id;
				 js.src = "//connect.facebook.net/en_US/sdk.js#xfbml=1&version=v2.6";
				 fjs.parentNode.insertBefore(js, fjs);
			   }(document, 'script', 'facebook-jssdk'));</script>
<!-- This site is optimized with the Yoast WordPress SEO plugin v1.6.1 - https://yoast.com/wordpress/plugins/seo/ -->
<meta content="noindex,follow" name="robots"/>
<link href="http://basketballintelligence.net/2013/05/" rel="canonical"/>
<link href="http://basketballintelligence.net/2013/05/page/2/" rel="next"/>
<meta content="en_US" property="og:locale"/>
<meta content="object" property="og:type"/>
<meta content="May 2013 - Basketball Intelligence" property="og:title"/>
<meta content="http://basketballintelligence.net/2013/05/" property="og:url"/>
<meta content="Basketball Intelligence" property="og:site_name"/>
<meta content="http://basketballintelligence.net/wp-content/uploads/2014/12/cropped-finals_11.png" property="og:image"/>
<!-- / Yoast WordPress SEO plugin. -->
<link href="//w.sharethis.com" rel="dns-prefetch"/>
<link href="//maxcdn.bootstrapcdn.com" rel="dns-prefetch"/>
<link href="//fonts.googleapis.com" rel="dns-prefetch"/>
<link href="//s.w.org" rel="dns-prefetch"/>
<link href="http://basketballintelligence.net/feed/" rel="alternate" title="Basketball Intelligence » Feed" type="application/rss+xml"/>
<link href="http://basketballintelligence.net/comments/feed/" rel="alternate" title="Basketball Intelligence » Comments Feed" type="application/rss+xml"/>
<script type="text/javascript">
			window._wpemojiSettings = {"baseUrl":"https:\/\/s.w.org\/images\/core\/emoji\/12.0.0-1\/72x72\/","ext":".png","svgUrl":"https:\/\/s.w.org\/images\/core\/emoji\/12.0.0-1\/svg\/","svgExt":".svg","source":{"concatemoji":"http:\/\/basketballintelligence.net\/wp-includes\/js\/wp-emoji-release.min.js?ver=5.3.2"}};
			!function(e,a,t){var r,n,o,i,p=a.createElement("canvas"),s=p.getContext&&p.getContext("2d");function c(e,t){var a=String.fromCharCode;s.clearRect(0,0,p.width,p.height),s.fillText(a.apply(this,e),0,0);var r=p.toDataURL();return s.clearRect(0,0,p.width,p.height),s.fillText(a.apply(this,t),0,0),r===p.toDataURL()}function l(e){if(!s||!s.fillText)return!1;switch(s.textBaseline="top",s.font="600 32px Arial",e){case"flag":return!c([127987,65039,8205,9895,65039],[127987,65039,8203,9895,65039])&&(!c([55356,56826,55356,56819],[55356,56826,8203,55356,56819])&&!c([55356,57332,56128,56423,56128,56418,56128,56421,56128,56430,56128,56423,56128,56447],[55356,57332,8203,56128,56423,8203,56128,56418,8203,56128,56421,8203,56128,56430,8203,56128,56423,8203,56128,56447]));case"emoji":return!c([55357,56424,55356,57342,8205,55358,56605,8205,55357,56424,55356,57340],[55357,56424,55356,57342,8203,55358,56605,8203,55357,56424,55356,57340])}return!1}function d(e){var t=a.createElement("script");t.src=e,t.defer=t.type="text/javascript",a.getElementsByTagName("head")[0].appendChild(t)}for(i=Array("flag","emoji"),t.supports={everything:!0,everythingExceptFlag:!0},o=0;o<i.length;o++)t.supports[i[o]]=l(i[o]),t.supports.everything=t.supports.everything&&t.supports[i[o]],"flag"!==i[o]&&(t.supports.everythingExceptFlag=t.supports.everythingExceptFlag&&t.supports[i[o]]);t.supports.everythingExceptFlag=t.supports.everythingExceptFlag&&!t.supports.flag,t.DOMReady=!1,t.readyCallback=function(){t.DOMReady=!0},t.supports.everything||(n=function(){t.readyCallback()},a.addEventListener?(a.addEventListener("DOMContentLoaded",n,!1),e.addEventListener("load",n,!1)):(e.attachEvent("onload",n),a.attachEvent("onreadystatechange",function(){"complete"===a.readyState&&t.readyCallback()})),(r=t.source||{}).concatemoji?d(r.concatemoji):r.wpemoji&&r.twemoji&&(d(r.twemoji),d(r.wpemoji)))}(window,document,window._wpemojiSettings);
		</script>
<style type="text/css">
img.wp-smiley,
img.emoji {
	display: inline !important;
	border: none !important;
	box-shadow: none !important;
	height: 1em !important;
	width: 1em !important;
	margin: 0 .07em !important;
	vertical-align: -0.1em !important;
	background: none !important;
	padding: 0 !important;
}
</style>
<link href="//maxcdn.bootstrapcdn.com/font-awesome/4.3.0/css/font-awesome.min.css?ver=5.3.2" id="simple-share-buttons-adder-font-awesome-css" media="all" rel="stylesheet" type="text/css"/>
<link href="http://basketballintelligence.net/wp-content/themes/superhero/style.css?ver=5.3.2" id="style-css" media="all" rel="stylesheet" type="text/css"/>
<link href="http://fonts.googleapis.com/css?family=Carrois+Gothic&amp;ver=5.3.2" id="superhero-carrois-gothic-css" media="all" rel="stylesheet" type="text/css"/>
<link href="http://basketballintelligence.net/wp-content/plugins/addthis/frontend/build/addthis_wordpress_public.min.css?ver=5.3.2" id="addthis_all_pages-css" media="all" rel="stylesheet" type="text/css"/>
<script id="st_insights_js" src="http://w.sharethis.com/button/st_insights.js?publisher=4d48b7c5-0ae3-43d4-bfbe-3ff8c17a8ae6&amp;product=simpleshare" type="text/javascript"></script>
<script src="http://basketballintelligence.net/wp-includes/js/jquery/jquery.js?ver=1.12.4-wp" type="text/javascript"></script>
<script src="http://basketballintelligence.net/wp-includes/js/jquery/jquery-migrate.min.js?ver=1.4.1" type="text/javascript"></script>
<script src="http://basketballintelligence.net/wp-content/themes/superhero/js/superhero.js?ver=5.3.2" type="text/javascript"></script>
<link href="http://basketballintelligence.net/wp-json/" rel="https://api.w.org/"/>
<link href="http://basketballintelligence.net/xmlrpc.php?rsd" rel="EditURI" title="RSD" type="application/rsd+xml"/>
<link href="http://basketballintelligence.net/wp-includes/wlwmanifest.xml" rel="wlwmanifest" type="application/wlwmanifest+xml"/>
<meta content="WordPress 5.3.2" name="generator"/>
<script id="mcjs">!function(c,h,i,m,p){m=c.createElement(h),p=c.getElementsByTagName(h)[0],m.async=1,m.src=i,p.parentNode.insertBefore(m,p)}(document,"script","https://chimpstatic.com/mcjs-connected/js/users/b2371955307f8826ade8aa973/d074eccdb5abd17800f80e9cf.js");</script> <style type="text/css">
			.site-title,
		.site-description {
			position: absolute;
			clip: rect(1px 1px 1px 1px); /* IE6, IE7 */
			clip: rect(1px, 1px, 1px, 1px);
		}
		</style>
<style id="custom-background-css" type="text/css">
body.custom-background { background-color: #f7f7f7; }
</style>
<script data-cfasync="false" type="text/javascript">if (window.addthis_product === undefined) { window.addthis_product = "wpp"; } if (window.wp_product_version === undefined) { window.wp_product_version = "wpp-6.2.6"; } if (window.addthis_share === undefined) { window.addthis_share = {}; } if (window.addthis_config === undefined) { window.addthis_config = {"data_track_clickback":true,"ui_atversion":300}; } if (window.addthis_plugin_info === undefined) { window.addthis_plugin_info = {"info_status":"enabled","cms_name":"WordPress","plugin_name":"Share Buttons by AddThis","plugin_version":"6.2.6","plugin_mode":"AddThis","anonymous_profile_id":"wp-e1123bbc47aadd91b2bbee06c04ad307","page_info":{"template":"archives","post_type":""},"sharing_enabled_on_post_via_metabox":false}; } 
                    (function() {
                      var first_load_interval_id = setInterval(function () {
                        if (typeof window.addthis !== 'undefined') {
                          window.clearInterval(first_load_interval_id);
                          if (typeof window.addthis_layers !== 'undefined' && Object.getOwnPropertyNames(window.addthis_layers).length > 0) {
                            window.addthis.layers(window.addthis_layers);
                          }
                          if (Array.isArray(window.addthis_layers_tools)) {
                            for (i = 0; i < window.addthis_layers_tools.length; i++) {
                              window.addthis.layers(window.addthis_layers_tools[i]);
                            }
                          }
                        }
                     },1000)
                    }());
                </script> <script async="async" data-cfasync="false" src="https://s7.addthis.com/js/300/addthis_widget.js#pubid=ra-54b00893220111c3" type="text/javascript"></script>
<!-- BEGIN ExactMetrics v5.3.9 Universal Analytics - https://exactmetrics.com/ -->
<script>
(function(i,s,o,g,r,a,m){i['GoogleAnalyticsObject']=r;i[r]=i[r]||function(){
	(i[r].q=i[r].q||[]).push(arguments)},i[r].l=1*new Date();a=s.createElement(o),
	m=s.getElementsByTagName(o)[0];a.async=1;a.src=g;m.parentNode.insertBefore(a,m)
})(window,document,'script','https://www.google-analytics.com/analytics.js','ga');
  ga('create', 'UA-56244307-1', 'auto');
  ga('send', 'pageview');
</script>
<!-- END ExactMetrics Universal Analytics -->
</head>
<body class="archive date custom-background group-blog elementor-default">
<div class="hfeed site" id="page">
<div id="masthead-wrap">
<header class="site-header" id="masthead" role="banner">
<hgroup>
<a href="http://basketballintelligence.net/" rel="home" title="Basketball Intelligence">
<img alt="" class="header-image" height="49" src="http://basketballintelligence.net/wp-content/uploads/2015/11/cropped-BI-logo2.png" width="409"/>
</a>
<h1 class="site-title"><a href="http://basketballintelligence.net/" rel="home" title="Basketball Intelligence">Basketball Intelligence</a></h1>
<h2 class="site-description">Todays Best NBA Reporting Analysis</h2>
</hgroup>
<nav class="site-navigation main-navigation" role="navigation">
<h1 class="assistive-text">Menu</h1>
<div class="assistive-text skip-link"><a href="#content" title="Skip to content">Skip to content</a></div>
<div class="menu-menu-1-container"><ul class="menu" id="menu-menu-1"><li class="menu-b-r menu-item menu-item-type-custom menu-item-object-custom menu-item-home menu-item-8" id="menu-item-8"><a href="http://basketballintelligence.net">Home</a></li>
<li class="menu-b-r menu-item menu-item-type-post_type menu-item-object-page menu-item-9679" id="menu-item-9679"><a href="http://basketballintelligence.net/about/">About</a></li>
<li class="menu-item menu-item-type-custom menu-item-object-custom menu-item-29349" id="menu-item-29349"><a href="https://bballwriters.com" rel="noopener noreferrer" target="_blank">The Basketball Writers</a></li>
</ul></div> </nav><!-- .site-navigation .main-navigation -->
<div class="clearfix"></div>
</header><!-- #masthead .site-header -->
</div><!-- #masthead-wrap -->
<div class="site-main" id="main">
<section class="content-area" id="primary">
<div class="site-content" id="content" role="main">
<header class="page-header">
<h1 class="page-title">
					Monthly Archives: <span>May 2013</span> </h1>
</header><!-- .page-header -->
<article class="post-472 post type-post status-publish format-standard hentry category-basketball-2 category-nba-2 category-professional-basketball category-uncategorized tag-david-west tag-indiana-pacers tag-joey-crawford tag-lance-stephenson tag-miami-heat tag-paul-george tag-roy-hibbert tag-zach-lowe" id="post-472">
<header class="entry-header">
<h1 class="entry-title"><a href="http://basketballintelligence.net/2013/05/30/how-have-the-pacers-done-it/" rel="bookmark">How have the Pacers done it?</a></h1>
<div class="entry-meta">
			Posted on <a href="http://basketballintelligence.net/2013/05/30/how-have-the-pacers-done-it/" rel="bookmark" title="7:33 am"><time class="entry-date" datetime="2013-05-30T07:33:52-07:00">May 30, 2013</time></a><span class="byline"> by <span class="author vcard"><a class="url fn n" href="http://basketballintelligence.net/author/bintel/" rel="author" title="View all posts by Basketball Intelligence">Basketball Intelligence</a></span></span> </div><!-- .entry-meta -->
</header><!-- .entry-header -->
<div class="entry-content">
<div class="at-above-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/30/how-have-the-pacers-done-it/"></div><p>From the great Zach Lowe at Grantland:</p>
<p>“….the Pacers have split four games with a team that had entered the series 45-3 in its prior 48 games. And they’ve done it for two reasons:</p>
<p>1. Indiana’s offense, which ranged from bad to mediocre all season, has sliced apart Miami at a rate of 111.3 points per 100 possessions — a mark that would have led the league in the regular season, and more than 10 points better than the Pacers managed in either of their first two series. The Pacers finished the season 19th in points per possession. Miami finished seventh in points allowed per possession.<br/>
<a name="more"></a><br/>
Not even Frank Vogel could have seen this coming, though his adjustments game-to-game have had a lot to do with Indiana’s shocking decimation of Miami’s coordinated chaos on defense. So have the following things: Indiana shooting 37 percent from 3-point range; Indiana getting to the line at a bonkers rate; Ray Allen’s inability to guard anyone other than, bizarrely, Paul George; and the Pacers rebounding 39.9 percent of their own misses, an offensive rebounding rate <a href="http://www.basketball-reference.com/play-index/tsl_finder.cgi?request=1&amp;match=single&amp;type=team_totals&amp;lg_id=NBA&amp;year_min=&amp;year_max=&amp;franch_id=&amp;c1stat=orb_pct&amp;c1comp=gt&amp;c1val=35.5&amp;c2stat=&amp;c2comp=gt&amp;c2val=&amp;c3stat=&amp;c3comp=gt&amp;c3val=&amp;c4stat=&amp;c4comp=gt&amp;c4val=&amp;order_by=orb_pct" target="_blank">that would be the highest in the history of the league</a> if any team actually pulled it off for an entire season. The Pacers couldn’t do that, of course. This is just a four-game sample size, but the comparison with season-long marks, while not useful in any deep way, indicates how badly the Pacers are dominating the glass.</p>
<p>2. Roy Hibbert is balling the hell out, on both ends of the floor. It’s astounding. Hibbert has been a very good player for a while, but this is something else. Hibbert has gone for 20 points and 10 boards in three of the four games in this series; he did that just seven times the entire regular season, and he went for 19 and nine in the one game this series that he missed the arbitrary round-number benchmark. And his defense hasn’t missed a beat, either.</p>
<p>The Pacers have gradually figured out how to get Hibbert the ball, and how to attack both Miami’s fronts and its hyper-aggressive trapping on pick-and-rolls.</p>
<p>…….</p>
<p>Underrated story: How often Miami is playing Chris Bosh and Andersen together in the second and fourth quarters. When the Heat are humming, they typically stay small, with Shane Battier or James at power forward, once Spoelstra removes Udonis Haslem midway through the first and third quarters. That’s not happening here, and Spoelstra’s shift back to big lineups is a statement of his respect for the Pacers’ size.</p>
<p>• In one stretch, George shifted over to Dwyane Wade in order to protect himself from further foul trouble. Wade has generally not been the sort of player an opposing team targets in this way. He looked semi-explosive in Game 3, and he did get to the line six times tonight — a rare bundle of free throws for Wade these days. But he struggled from the floor and looked dispirited.</p>
<p>• This series is almost exactly even after four games, which means Miami has lit up</p>
<p>Indiana’s defense at about the same rate as the Pacers have torn apart Miami’s vaunted D. The Heat have continued to hurt Indiana by spacing Hibbert’s man way out along the baseline, though Hibbert has been quicker to help at the rim since Game 1. They’ve also found some pet sets they like in this series.</p>
<p>But that Pacers defense, like any elite unit, is capable of long stretches of stingy play, and they unleashed a 3:22 masterpiece after the Heat had taken a 60-54 lead in the third quarter. There’s nothing fancy about what Indiana does. They’re huge at every position, they have a lot of individual talent, and they work well together as a team. They understand Vogel’s priorities, and they rarely make mistakes.</p>
<p>……</p>
<p>The Pacers have also done remarkably well taking away any air space for Allen, Battier, and Miami’s other spot-up guys. The Heat have tried to free up Battier by using him as the screener in pick-and-pops, but West has done well cutting off the pick-and-roll drive and scampering back to Battier before he can get off a shot. “We want to force them to make plays off the bounce,” West said after the game.”</p>
<p>Lots more to this story by Zach, including illustrative videos and his take on Lance Stephenson at <b>http://tinyurl.com/q6pn78y</b></p>
<!-- AddThis Advanced Settings above via filter on the_content --><!-- AddThis Advanced Settings below via filter on the_content --><!-- AddThis Advanced Settings generic via filter on the_content --><!-- AddThis Share Buttons above via filter on the_content --><!-- AddThis Share Buttons below via filter on the_content --><div class="at-below-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/30/how-have-the-pacers-done-it/"></div><!-- AddThis Share Buttons generic via filter on the_content --> </div><!-- .entry-content -->
<footer class="entry-meta">
<span class="cat-links">
				Posted in <a href="http://basketballintelligence.net/category/basketball-2/" rel="category tag">basketball</a>, <a href="http://basketballintelligence.net/category/nba-2/" rel="category tag">NBA</a>, <a href="http://basketballintelligence.net/category/professional-basketball/" rel="category tag">professional basketball</a>, <a href="http://basketballintelligence.net/category/uncategorized/" rel="category tag">Uncategorized</a> </span>
<span class="sep"> | </span>
<span class="tags-links">
				Tagged <a href="http://basketballintelligence.net/tag/david-west/" rel="tag">David West</a>, <a href="http://basketballintelligence.net/tag/indiana-pacers/" rel="tag">Indiana Pacers</a>, <a href="http://basketballintelligence.net/tag/joey-crawford/" rel="tag">Joey Crawford</a>, <a href="http://basketballintelligence.net/tag/lance-stephenson/" rel="tag">Lance Stephenson</a>, <a href="http://basketballintelligence.net/tag/miami-heat/" rel="tag">Miami Heat</a>, <a href="http://basketballintelligence.net/tag/paul-george/" rel="tag">Paul George</a>, <a href="http://basketballintelligence.net/tag/roy-hibbert/" rel="tag">Roy Hibbert</a>, <a href="http://basketballintelligence.net/tag/zach-lowe/" rel="tag">Zach Lowe</a> </span>
</footer><!-- .entry-meta -->
</article><!-- #post-## -->
<article class="post-461 post type-post status-publish format-standard hentry category-basketball-2 category-nba-2 category-professional-basketball category-uncategorized tag-david-west tag-flopping tag-ken-berger tag-lebron-james tag-nba-nba-playoffs" id="post-461">
<header class="entry-header">
<h1 class="entry-title"><a href="http://basketballintelligence.net/2013/05/29/flopping-in-the-playoffs/" rel="bookmark">Flopping in the Playoffs</a></h1>
<div class="entry-meta">
			Posted on <a href="http://basketballintelligence.net/2013/05/29/flopping-in-the-playoffs/" rel="bookmark" title="12:09 pm"><time class="entry-date" datetime="2013-05-29T12:09:41-07:00">May 29, 2013</time></a><span class="byline"> by <span class="author vcard"><a class="url fn n" href="http://basketballintelligence.net/author/bintel/" rel="author" title="View all posts by Basketball Intelligence">Basketball Intelligence</a></span></span> </div><!-- .entry-meta -->
</header><!-- .entry-header -->
<div class="entry-content">
<div class="at-above-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/29/flopping-in-the-playoffs/"></div><p>From Ken Berger at CBSSports.com:</p>
<p>“David West leaned against the basket stanchion on Monday and paused for a long time after I asked my question. His eyes gazed toward the roof, and he smiled.</p>
<p>My question: How well has the NBA’s anti-flopping policy worked this season?</p>
<p>His answer: Not as bad as the answer from <a href="http://www.cbssports.com/nba/players/playerpage/400553/lebron-james">LeBron James</a>.</p>
<p>First, West’s answer: “I know fining a guy doesn’t really take away a foul at a crucial moment of a game. It’s so hard. It’s a judgment call by the refs, and sometimes guys get rewarded for it.”</p>
<aside>Sometimes? The NBA during the postseason has been as flop-tastic as ever. Next, I’m half expecting to read that <a href="http://www.cbssports.com/nba/players/playerpage/240304/tony-parker">Tony Parker</a> actually <em>did</em> leave his hotel room that night in Memphis, but never made it to the fancy restaurant in question because someone brushed against him on the sidewalk and he landed in a sewer drain three blocks away.</aside>
<p>Now for the disturbing part: LeBron James’ answer to my question about flopping on the practice day between Games 3 and 4 of the Eastern Conference finals. Get a load of this.</p>
<p>“It hasn’t been a problem for many guys at all,” James said. “I don’t really pay too much attention to it. I think it’s been good, I guess.”</p>
<p>And since we’re guessing here, I guess LeBron was talking about the policy itself, not the fact that it has proved to be woefully ineffective at stopping the bane of basketball, the trickery and foolishness that are ruining the games.</p>
<p>“It’s year one, so you’re not just going to go cold turkey,” James said. “Guys have been accustomed to doing it for years, and it’s not even a bad thing. You’re just trying to get the advantage. Any way you can get the advantage over an opponent to help your team win, then so be it.”</p>
<p>Here, I was floored. You would’ve thought LeBron had grazed my shoulder, because I almost went flailing 30 feet across the Pacers’ practice court when he said that. If I’d done that while wearing a <a href="http://www.cbssports.com/nba/players/playerpage/139066/manu-ginobili">Manu Ginobili</a> jersey, of course, nobody would’ve given a second thought.</p>
<aside>
<header>The biggest star, the most recognizable face of the NBA, says flopping is “not even a bad thing?” He says that “any way you can get the advantage over an opponent … then so be it?”</header>
</aside>
<p>Well then, that says everything we need to know — everything we already understood — about the ineffectiveness of the NBA’s steps to eradicate flopping.”</p>
<p>Read the rest of Ken’s story here:  http://www.cbssports.com/nba/story/22314389/playoffs-as-floptastic-as-ever-and-nba-doing-little-to-curb-problem</p>
<div>And, an historical perspective:</div>
<div></div>
<div>http://www.grantland.com/blog/the-triangle/post/_/id/63835/flopping-in-the-nba-a-history-of-nonviolence</div>
<div></div>
<p>Some other recent stories of note:</p>
<div>Fans React to Magic dropping Matt Guokas:</div>
<div></div>
<div id="yui_3_7_2_1_1369837916512_16169">http://www.orlandosentinel.com/sports/blogs/magic-basketblog/os-mbb-fans-media-react-to-matt-guokas-news-20130524,0,2184736.post</div>
<div></div>
<div></div>
<div>8-Second Rule isn’t that complicated:</div>
<div></div>
<div id="yui_3_7_2_1_1369837916512_16170">http://nesn.com/2013/05/nbas-seemingly-contradictory-shot-clock-eight-second-violation-rule-really-not-that-complicated/</div>
<div></div>
<div></div>
<div>Jerry West’s thoughts on turning 75:</div>
<div></div>
<div id="yui_3_7_2_1_1369837916512_16173">http://www.nba.com/2013/news/features/scott_howard_cooper/05/28/jerry-west-thoughts-on-turning-75/</div>
<div></div>
<div></div>
<div>Memphis Grizzlies, the Cable Drama:</div>
<div></div>
<div>http://espn.go.com/blog/truehoop/post/_/id/58696/memphis-grizzlies-the-cable-drama</div>
<div></div>
<div></div>
<p> </p>
<!-- AddThis Advanced Settings above via filter on the_content --><!-- AddThis Advanced Settings below via filter on the_content --><!-- AddThis Advanced Settings generic via filter on the_content --><!-- AddThis Share Buttons above via filter on the_content --><!-- AddThis Share Buttons below via filter on the_content --><div class="at-below-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/29/flopping-in-the-playoffs/"></div><!-- AddThis Share Buttons generic via filter on the_content --> </div><!-- .entry-content -->
<footer class="entry-meta">
<span class="cat-links">
				Posted in <a href="http://basketballintelligence.net/category/basketball-2/" rel="category tag">basketball</a>, <a href="http://basketballintelligence.net/category/nba-2/" rel="category tag">NBA</a>, <a href="http://basketballintelligence.net/category/professional-basketball/" rel="category tag">professional basketball</a>, <a href="http://basketballintelligence.net/category/uncategorized/" rel="category tag">Uncategorized</a> </span>
<span class="sep"> | </span>
<span class="tags-links">
				Tagged <a href="http://basketballintelligence.net/tag/david-west/" rel="tag">David West</a>, <a href="http://basketballintelligence.net/tag/flopping/" rel="tag">Flopping</a>, <a href="http://basketballintelligence.net/tag/ken-berger/" rel="tag">Ken Berger</a>, <a href="http://basketballintelligence.net/tag/lebron-james/" rel="tag">Lebron James</a>, <a href="http://basketballintelligence.net/tag/nba-nba-playoffs/" rel="tag">NBA. NBA Playoffs</a> </span>
</footer><!-- .entry-meta -->
</article><!-- #post-## -->
<article class="post-454 post type-post status-publish format-standard hentry category-basketball-2 category-nba-2 category-professional-basketball category-uncategorized tag-david-aldridge tag-david-west tag-donnie-walsh tag-indiana-pacers tag-lance-stephenson tag-larry-bird tag-miami-heat tag-paul-george tag-roy-hibbert" id="post-454">
<header class="entry-header">
<h1 class="entry-title"><a href="http://basketballintelligence.net/2013/05/28/larry-bird-architect-of-pacers-rise/" rel="bookmark">Larry Bird, architect of Pacers’ rise</a></h1>
<div class="entry-meta">
			Posted on <a href="http://basketballintelligence.net/2013/05/28/larry-bird-architect-of-pacers-rise/" rel="bookmark" title="7:08 am"><time class="entry-date" datetime="2013-05-28T07:08:42-07:00">May 28, 2013</time></a><span class="byline"> by <span class="author vcard"><a class="url fn n" href="http://basketballintelligence.net/author/bintel/" rel="author" title="View all posts by Basketball Intelligence">Basketball Intelligence</a></span></span> </div><!-- .entry-meta -->
</header><!-- .entry-header -->
<div class="entry-content">
<div class="at-above-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/28/larry-bird-architect-of-pacers-rise/"></div><h1>Bird, architect of Pacers’ rise, chooses anonymity over acclaim</h1>
<p>(Excerpted from a story by David Aldridge at NBA.com.  Read the whole story here: <b>http://tinyurl.com/p42pl2h</b></p>
<p><strong> </strong> The man who put the Indiana Pacers together is, this morning, likely at his compound in Indiana, maybe fishing on one of his lakes. Or maybe he’s tinkering about at the main house.</p>
<p>It shouldn’t surprise you that he lives in one of the least densely populated counties in the state. He is incommunicado with all but a few people, the way Greta Garbo used to be, the way Howard Hughes used to be, the way a lot of famous people used to be before Twitter, Facebook, Tumblr and all the rest.</p>
<p>If you do not seek fame, if you do not burn to walk on the red carpet or be mentioned in US Weekly, you can keep fame at arm’s length. It is still possible to be a private person, to go away and stay away.</p>
<div>
<div>
<table>
<tbody>
<tr>
<td colspan="2"></td>
</tr>
<tr>
<td></td>
<td></td>
</tr>
</tbody>
</table>
</div>
</div>
<p>Larry Bird stays away.</p>
<p>He is a rumor, a whisper, accessible only to a handful of media people he knows and trusts, like the great Jackie MacMullen, with whom he wrote a couple of books and that he got to know in Boston. He is available, of course, with his close circle of friends and family. But to most of the world, he has disappeared.</p>
<p>He was not in Indianapolis on Sunday, when they ran the 500 as they always do on Memorial Day Weekend, and when the Pacers played in the Eastern Conference finals, which they had not done since 2004.</p>
<p>The team that Bird put together looks for all intents and purposes like the only team in the East capable of giving LeBron James and the Heat a battle for the next few seasons. Indiana is down 2-1 after getting vaporized Sunday night, but no one with a brain is writing the Pacers off, not after they went into Miami, gave away Game 1 and took away Game 2. The Pacers still have a chance to get back to The Finals for the first time since Bird coached them in 2000, because Bird saw things that others didn’t see on the basketball court.</p>
<p>The man Bill Fitch famously called “Kodak” because of his extraordinary vision has built a championship contender only to vanish himself during its apex.</p>
<p>It was Bird that traded for Roy Hibbert on Draft night in 2008, a center prospect considered so slow and so heavy that other teams picking just ahead of Toronto — which took Hibbert with the 17th pick and then sent him to Indiana — didn’t even have Hibbert on their boards.</p>
<p>It was Bird who took a little-regarded forward from Fresno State named Paul George with the 10th pick in 2010, a player that was dismissed by a lot of personnel types because of the relative lack of competition the Bulldogs faced, in part because people thought George was a little soft.</p>
<p>It was Bird that convinced David West to sign in Indiana, and not Boston, as a free agent two years ago … even though the Celtics were offering more money. It was Bird that used a second-rounder in 2010 on Lance Stephenson, whose troubled past gave most NBA teams the willies. And it was Bird that didn’t hesitate to promote a then 37-year-old assistant coach with no prior head coaching experience — though he was a star point guard for Juniata (Penn.) College — to be his coach in 2011, replacing Jim O’Brien.</p>
<p>But just when the Pacers finally turned the corner and began to win back fans who had soured on the team after the Brawl of Auburn Hills in 2004 — and other off-court incidents that followed — Bird walked away. He resigned as team president last summer, saying he would take at least a year off to deal with health issues. Donnie Walsh came back from New York to take Bird’s job, with ex-Blazers general manager Kevin Pritchard doing the grind work as new GM.</p>
<div>
<div>Your browser does not support iframes.</div>
</div>
<p>There are signals that Bird may want to return to the presidency next season. Walsh has indicated he would have no problem stepping aside. But for now, Bird’s absence has created the oddest dynamic: the people now in charge of keeping the Pacers together had almost nothing to do with bringing them together.</p>
<p>“I’ve told him, he’s weird to do that,” Walsh said Sunday. “But he did that when he was my coach and we went to The Finals with him, and he resigned right after that. I would ask him, ‘What’s your dynamic here? Why would you leave now?’ But I’ve learned to never psychoanalyze Bird.”</p>
<!-- AddThis Advanced Settings above via filter on the_content --><!-- AddThis Advanced Settings below via filter on the_content --><!-- AddThis Advanced Settings generic via filter on the_content --><!-- AddThis Share Buttons above via filter on the_content --><!-- AddThis Share Buttons below via filter on the_content --><div class="at-below-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/28/larry-bird-architect-of-pacers-rise/"></div><!-- AddThis Share Buttons generic via filter on the_content --> </div><!-- .entry-content -->
<footer class="entry-meta">
<span class="cat-links">
				Posted in <a href="http://basketballintelligence.net/category/basketball-2/" rel="category tag">basketball</a>, <a href="http://basketballintelligence.net/category/nba-2/" rel="category tag">NBA</a>, <a href="http://basketballintelligence.net/category/professional-basketball/" rel="category tag">professional basketball</a>, <a href="http://basketballintelligence.net/category/uncategorized/" rel="category tag">Uncategorized</a> </span>
<span class="sep"> | </span>
<span class="tags-links">
				Tagged <a href="http://basketballintelligence.net/tag/david-aldridge/" rel="tag">David Aldridge</a>, <a href="http://basketballintelligence.net/tag/david-west/" rel="tag">David West</a>, <a href="http://basketballintelligence.net/tag/donnie-walsh/" rel="tag">Donnie Walsh</a>, <a href="http://basketballintelligence.net/tag/indiana-pacers/" rel="tag">Indiana Pacers</a>, <a href="http://basketballintelligence.net/tag/lance-stephenson/" rel="tag">Lance Stephenson</a>, <a href="http://basketballintelligence.net/tag/larry-bird/" rel="tag">Larry Bird</a>, <a href="http://basketballintelligence.net/tag/miami-heat/" rel="tag">Miami Heat</a>, <a href="http://basketballintelligence.net/tag/paul-george/" rel="tag">Paul George</a>, <a href="http://basketballintelligence.net/tag/roy-hibbert/" rel="tag">Roy Hibbert</a> </span>
</footer><!-- .entry-meta -->
</article><!-- #post-## -->
<article class="post-444 post type-post status-publish format-standard hentry category-basketball-2 category-nba-2 category-professional-basketball category-uncategorized tag-kevin-arnovitz tag-memphis-grizzlies tag-mike-conley tag-nba tag-nba-playoffs tag-zach-randolph" id="post-444">
<header class="entry-header">
<h1 class="entry-title"><a href="http://basketballintelligence.net/2013/05/27/grizzlies-shooting-pains/" rel="bookmark">Grizzlies’ “Shooting Pains”</a></h1>
<div class="entry-meta">
			Posted on <a href="http://basketballintelligence.net/2013/05/27/grizzlies-shooting-pains/" rel="bookmark" title="6:44 am"><time class="entry-date" datetime="2013-05-27T06:44:53-07:00">May 27, 2013</time></a><span class="byline"> by <span class="author vcard"><a class="url fn n" href="http://basketballintelligence.net/author/bintel/" rel="author" title="View all posts by Basketball Intelligence">Basketball Intelligence</a></span></span> </div><!-- .entry-meta -->
</header><!-- .entry-header -->
<div class="entry-content">
<div class="at-above-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/27/grizzlies-shooting-pains/"></div><p>Excerpt from Kevin Arnovitz’ commentary at ESPN.com:  (To see the whole piece, go to http://espn.go.com/nba/story/_/id/9312302/memphis-grizzlies-approaching-exit-playoffs-shooting-just-there)</p>
<h1>Shooting pains at heart of Griz’s woe</h1>
<p><cite>“….(</cite>W)ith little outside shooting to open up the paint, Zach Randolph has found things quite congested.</p>
<hr/>
<p>The Grizzlies never lack for effort, and it’s one of the primary reasons they’ve hung around until Memorial Day weekend, outlasting teams with greater offensive talent and lesser flaws. But there are times when the gears grind and precision erodes, when it looks like the Grizzlies will never score another basket.</p>
<p>“We’re not efficient with what effort we’re getting,” Hollins said after the game. “That’s where we have to be better.”</p>
<p>This has never been a simple formula for the Grizzlies. Since the current core was assembled, the Grizzlies have never had a platoon of pure, knockdown shooters. Conley has graduated from iffy to proficient. <a href="http://espn.go.com/nba/player/_/id/3450/oj-mayo">O.J. Mayo</a> was the team’s leading 3-point shooter the previous two seasons in shots made, but he hit at an unexceptional 36.4 percent clip both seasons. None of the regulars this season touched 40 percent.</p>
<p>A team whose most lethal shooter from long range is Quincy Pondexter probably has no business in a conference finals, and players, coaches and team execs have all acknowledged these shortcomings. And it’s this self-realization, a firm understanding of what the team can and can’t accomplish as a unit that’s made the Grizzlies so good over the past few months. Self-awareness is a powerful thing, and it has endowed Memphis with a true survival instinct. The Grizzlies have managed to find enough table scraps as bullies on the offensive glass and predators in the post. They get periodic bursts of confidence from Conley and sneaky play behind the defenses from Allen.</p>
<p>Yet a team can gnaw at the margins for only so long before the bone is picked clean. The Spurs’ big men keep two feet in the paint, while their lanky wing defenders swoop in from the baseline and load up. Routine post entries into Gasol and Randolph are messy affairs, and when the ball finds its way into the paint, the help is almost always on time.</p>
<p>In today’s NBA, long-range shooting and speed are the most reliable antidotes to this brand of defense, but this is where the Grizzlies are in short supply, and it’s finally caught up to them. These are shortcomings Memphis will undoubtedly address on the wing this offseason, but until then there simply isn’t enough firepower to burst through the Spurs’ barricade.</p>
<p>Traditionally, teams that lose the first three games of an NBA playoff series are buried beneath history. No team has ever come back from a 3-0 series deficit, and the well-seasoned, poised Spurs are an unlikely victim to defy probabilities. But the Grizzlies still speak hopefully, even in their self-critiques.</p>
<p>“We’ve got to do other stuff, get to our second and third options and get the ball upcourt fast — that’s something we’ve got to work on,” <a href="http://espn.go.com/nba/player/_/id/1017/zach-randolph">Zach Randolph</a> said after the game, as if these tasks were long-term projects.</p>
<p>“We’ve been learning for the past few games,” Gasol said with all the confidence of a guy who expects to see this learning process to fruition. A moment later, Gasol used the future simple tense: “We’re going to have to communicate more, like we always do.”</p>
<p>“It’s a good learning lesson,” Conley said at the podium, echoing Gasol’s sentiments. “There’s no group of guys I’d rather be with down 3-0 to try to fight back into it.”</p>
<p>There’s no pleasant way to cede defeat, but the Grizzlies have been playing with house money for the better part of a month. That their key young players can own up to their flaws yet still speak in the language of the future bodes well for Memphis.”</p>
<!-- AddThis Advanced Settings above via filter on the_content --><!-- AddThis Advanced Settings below via filter on the_content --><!-- AddThis Advanced Settings generic via filter on the_content --><!-- AddThis Share Buttons above via filter on the_content --><!-- AddThis Share Buttons below via filter on the_content --><div class="at-below-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/27/grizzlies-shooting-pains/"></div><!-- AddThis Share Buttons generic via filter on the_content --> </div><!-- .entry-content -->
<footer class="entry-meta">
<span class="cat-links">
				Posted in <a href="http://basketballintelligence.net/category/basketball-2/" rel="category tag">basketball</a>, <a href="http://basketballintelligence.net/category/nba-2/" rel="category tag">NBA</a>, <a href="http://basketballintelligence.net/category/professional-basketball/" rel="category tag">professional basketball</a>, <a href="http://basketballintelligence.net/category/uncategorized/" rel="category tag">Uncategorized</a> </span>
<span class="sep"> | </span>
<span class="tags-links">
				Tagged <a href="http://basketballintelligence.net/tag/kevin-arnovitz/" rel="tag">Kevin Arnovitz</a>, <a href="http://basketballintelligence.net/tag/memphis-grizzlies/" rel="tag">Memphis Grizzlies</a>, <a href="http://basketballintelligence.net/tag/mike-conley/" rel="tag">Mike Conley</a>, <a href="http://basketballintelligence.net/tag/nba/" rel="tag">NBA</a>, <a href="http://basketballintelligence.net/tag/nba-playoffs/" rel="tag">NBA Playoffs</a>, <a href="http://basketballintelligence.net/tag/zach-randolph/" rel="tag">Zach Randolph</a> </span>
</footer><!-- .entry-meta -->
</article><!-- #post-## -->
<article class="post-432 post type-post status-publish format-standard hentry category-basketball-2 category-nba-2 category-professional-basketball category-uncategorized tag-blocked-shots tag-deflections tag-frank-vogel tag-nba tag-nba-defense tag-nba-playoffs tag-per tag-rim-protector tag-roy-hibbert tag-sportvu" id="post-432">
<header class="entry-header">
<h1 class="entry-title"><a href="http://basketballintelligence.net/2013/05/26/roy-hibbert-the-protector/" rel="bookmark">“Roy Hibbert, the Protector”</a></h1>
<div class="entry-meta">
			Posted on <a href="http://basketballintelligence.net/2013/05/26/roy-hibbert-the-protector/" rel="bookmark" title="10:45 am"><time class="entry-date" datetime="2013-05-26T10:45:47-07:00">May 26, 2013</time></a><span class="byline"> by <span class="author vcard"><a class="url fn n" href="http://basketballintelligence.net/author/bintel/" rel="author" title="View all posts by Basketball Intelligence">Basketball Intelligence</a></span></span> </div><!-- .entry-meta -->
</header><!-- .entry-header -->
<div class="entry-content">
<div class="at-above-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/26/roy-hibbert-the-protector/"></div><p>What follows below is excerpted from a piece by Kirk Goldsberry at Grantland.  To see it in its entirety (with visuals included as well), go here:  http://www.grantland.com/blog/the-triangle/post/_/id/63158/courtvision-roy-hibbert-the-protector</p>
<p>“…(W)hen Indiana signed Hibbert to a max deal last summer, many NBA observers questioned the decision. From a statistical point of view, Hibbert’s numbers seemed more ordinary than elite. Beyond the statistics, he lacked the Q Score that usually goes along with the league’s highest-paid superstars. But here we are 10 months later and Hibbert has his team in the NBA’s final four, and he is a major reason for Indiana’s reputation as one of the best defenses in the league. For those who value both sides of the basketball court, he has more than proven his worth.</p>
<p>In our <em>SportsCenter</em> culture, which loves to watch and rewatch dunks, corner 3s, and buzzer-beaters, it’s easy to distort basketball value. When it comes to assessing on-court performance, a vast majority of influential basketball actions aren’t compatible with either highlights or box scores. Things like deflected passes, altered shots, prevented shots, perfect defensive rotations, and impeccable screens go largely unnoticed by the masses, undocumented by the highlight shows, and undermeasured by even the nerdiest basketball bookkeepers. As a result, our spreadsheets are missing a bunch of meaningful columns, and in turn even our most advanced player evaluations are incomplete.</p>
<p>The inability to justify Hibbert’s contract with our current stats says more about our current stats than it does about Hibbert’s value.</p>
<hr/>
<p>I talked with Hibbert about interior defense on Thursday, the day after Indiana’s heartbreaking Game 1 loss in Miami. He told me he finds that blocked shots — the stat that most of us cite when discussing players like him — tell only part of the story. Hibbert thinks he influences more than just the shots he blocks: “There are a lot of shots that I don’t block, but I change shots, and obviously people don’t take that into account. It might not show up in the box score, but people around my team know what I bring.”</p>
<p>Blocked shots are easy to detect and convenient to count, but they reveal only so much about a defensive performance. Hibbert had five blocks in Game 6 of the Knicks series, but what does that actually tell us? On a literal level, his hands interrupted the flight of five field goal attempts. So what? Aside from that insane block of Anthony, how many other shots did he prevent from even happening? How many possessions did he change? Surely, his presence influenced more than five shots during the game, but it’s that “5” in the box score that serves as the quantification for defensive contribution.</p>
<p>This season the Pacers defense, with Hibbert as its anchor, held opponents to 96.6 points per 100 possessions, the lowest mark in the league. It’s no coincidence that it also protected the basket better than anybody. Indiana was the only team whose opponents missed more close-range shots than they made. On average, Indiana’s opponents shot just 49 percent within 8 feet of the basket, well below the league average of 56 percent. A lot of this has to do with Hibbert.</p>
<p>Using emerging types of performance data, we can evaluate interior defense in new ways. Player-tracking systems, like the SportVU system currently installed in 15 NBA arenas, enable a much richer perspective than the conventional mode of simply summing up disparate event types like blocked shots. While Hibbert finished third in blocks per game this season, averaging 2.5, an analysis of his defense using SportVU data reveals that his influence is far more profound.</p>
<p>Hibbert significantly reduces his opponents’ overall scoring efficiency on a nightly basis. I evaluated a set of thousands of NBA close-range shots in which an NBA big man was protecting the basket. These were shots from the 2012-13 regular season in which a qualifying interior defender was within 5 feet of the rim and also within 5 feet of the shot location.</p>
<p>In such cases, the opponents made 48 percent of their shots. When Hibbert was protecting the basket, however, the number dropped to 38 percent. Only one player in the NBA reduced close-range shooting efficiency more than Hibbert; of course, that was Larry Sanders, who held opponents to a ghastly 32 percent. For context, both Marc Gasol and Tyson Chandler — the last two winners of the NBA Defensive Player of the Year Award — held opponents to a respectable 44 percent.</p>
<p>When Hibbert is protecting the basket, opponents’ <em>close-range</em> shots go in about as much as an average NBA <em>midrange</em> shot. This is incredible. If we’re not factoring in this kind of precipitous drop in opponents’ scoring efficiency when Hibbert is protecting the rim (which we’re not), then we’re not doing a good job assessing his impact or value as an NBA player.</p>
<p>Hibbert told me that since entering the NBA he’s <a href="http://www.grantland.com/story/_/id/7881895/how-indiana-pacers-all-star-center-roy-hibbert-avoided-becoming-stiff" target="new">become more disciplined</a>: “When I was a rookie, I tried to block everything. I led the league in fouls per minute. Since then I’ve learned <a href="http://www.nba.com/video/channels/originals/2012/01/24/bballops_11_403.nba/index.html" target="_blank">verticality</a>, and that’s one thing that helps me both protect the paint and stay in the game as well.”</p>
<p>Earlier in the season after a win at Chicago, <a href="http://hangtime.blogs.nba.com/2012/12/05/pacers-hibbert-busts-deng-bulls-late-with-law-of-verticality/" target="_blank">Vogel explained</a>: “He’s the biggest reason why we lead the league in field goal defense. He’s the best in the league at exercising the fundamental of verticality. Using his legs, getting off his feet and making a legal defensive play, and earning a no-call. You’re allowed to jump straight up, no matter where you are, and absorb contact. When he learned that and went away from trying to draw charges, like he was earlier in his career, he went from not being able to stay on the court to being one of the best defensive centers in the NBA.”</p>
<p>Hibbert also habitually watches film, studying his opponents’ shooting tendencies. “I make sure I know who likes to shoot with the right hand, and who likes to shoot with their left hand.”</p>
<p>His former student team manager at Georgetown and his current development consultant, Justin Zormelo, echoed this idea: “He’s the first person to let me know if I’m an hour late sending him stuff to prepare. He made it a goal to be the best defensive player in the league a couple years ago and he’s been preparing himself and working at it every day since.”</p>
<hr/>
<p>Despite the rapidly increasing importance of the 3-pointer in the NBA, good shots close to the basket are still the best shots on the court. No matter what level you’re playing at, if you can’t protect the basket, you are going to lose. Players who <em>can</em> protect are extremely valuable, even if we still have problems detecting and quantifying that value.</p>
<p>At 26 years old, Hibbert is one of the most important players on a team in the conference finals… Still, Hibbert’s Player Efficiency Rating in the regular season was 17.3, placing him just below Amir Johnson and just above Kosta Koufos. With all due respect to those guys, something is awry here. Despite so-so defensive reputations, guys like (David) Lee and (Blake) Griffin are All-NBA performers with lucrative endorsement deals. Sure, they might “get you 20 and 10 every night,” but they might also give up 25 and 10 in the process. One problem is that the 20 and 10 are duly noted and factored into PER while the 25 and 10 allowed remain mostly unmeasured.</p>
<p>While it’s enticing to pat ourselves on the back and say basketball has entered its “advanced metrics” era, our inability to properly quantify the impact of players like Hibbert reminds us that we are still in the Stone Age.”</p>
<footer>
<section>
<div id="facebook-recommend"></div>
</section>
</footer>
<aside>
<div id="facebook-comments"></div>
</aside>
<div id="col-right"></div>
<!-- AddThis Advanced Settings above via filter on the_content --><!-- AddThis Advanced Settings below via filter on the_content --><!-- AddThis Advanced Settings generic via filter on the_content --><!-- AddThis Share Buttons above via filter on the_content --><!-- AddThis Share Buttons below via filter on the_content --><div class="at-below-post-arch-page addthis_tool" data-url="http://basketballintelligence.net/2013/05/26/roy-hibbert-the-protector/"></div><!-- AddThis Share Buttons generic via filter on the_content --> </div><!-- .entry-content -->
<footer class="entry-meta">
<span class="cat-links">
				Posted in <a href="http://basketballintelligence.net/category/basketball-2/" rel="category tag">basketball</a>, <a href="http://basketballintelligence.net/category/nba-2/" rel="category tag">NBA</a>, <a href="http://basketballintelligence.net/category/professional-basketball/" rel="category tag">professional basketball</a>, <a href="http://basketballintelligence.net/category/uncategorized/" rel="category tag">Uncategorized</a> </span>
<span class="sep"> | </span>
<span class="tags-links">
				Tagged <a href="http://basketballintelligence.net/tag/blocked-shots/" rel="tag">blocked shots</a>, <a href="http://basketballintelligence.net/tag/deflections/" rel="tag">deflections</a>, <a href="http://basketballintelligence.net/tag/frank-vogel/" rel="tag">Frank Vogel</a>, <a href="http://basketballintelligence.net/tag/nba/" rel="tag">NBA</a>, <a href="http://basketballintelligence.net/tag/nba-defense/" rel="tag">NBA defense</a>, <a href="http://basketballintelligence.net/tag/nba-playoffs/" rel="tag">NBA Playoffs</a>, <a href="http://basketballintelligence.net/tag/per/" rel="tag">PER</a>, <a href="http://basketballintelligence.net/tag/rim-protector/" rel="tag">Rim Protector</a>, <a href="http://basketballintelligence.net/tag/roy-hibbert/" rel="tag">Roy Hibbert</a>, <a href="http://basketballintelligence.net/tag/sportvu/" rel="tag">SportVu</a> </span>
</footer><!-- .entry-meta -->
</article><!-- #post-## -->
<nav class="site-navigation paging-navigation" id="nav-below" role="navigation">
<h1 class="assistive-text">Post navigation</h1>
<div class="nav-previous"><a href="http://basketballintelligence.net/2013/05/page/2/"><span class="meta-nav">←</span> Older posts</a></div>
</nav><!-- #nav-below -->
</div><!-- #content -->
</section><!-- #primary -->
<div class="widget-area" id="secondary" role="complementary">
<aside class="widget widget_search" id="search-2"><form action="http://basketballintelligence.net/" id="searchform" method="get" role="search">
<label class="assistive-text" for="s">Search</label>
<input class="field" id="s" name="s" placeholder="Search …" type="text" value=""/>
<input class="submit" id="searchsubmit" name="submit" type="submit" value="Search"/>
</form>
</aside><aside class="widget widget_text" id="text-6"> <div class="textwidget"><!-- Begin MailChimp Signup Form -->
<link href="//cdn-images.mailchimp.com/embedcode/slim-10_7.css" rel="stylesheet" type="text/css"/>
<style type="text/css">
	#mc_embed_signup{background:#fff; clear:left; font:14px Helvetica,Arial,sans-serif;  width:220px;}
	/* Add your own MailChimp form style overrides in your site stylesheet or in this style block.
	   We recommend moving this block and the preceding CSS link to the HEAD of your HTML file. */
</style>
<div id="mc_embed_signup">
<form action="//Net.us9.list-manage.com/subscribe/post?u=b2371955307f8826ade8aa973&amp;id=3d22d9fd4e" class="validate" id="mc-embedded-subscribe-form" method="post" name="mc-embedded-subscribe-form" novalidate="" target="_blank">
<div id="mc_embed_signup_scroll">
<label for="mce-EMAIL">Subscribe By Email</label>
<input class="email" id="mce-EMAIL" name="EMAIL" placeholder="email address" required="" type="email" value=""/>
<!-- real people should not fill this in and expect good things - do not remove this or risk form bot signups-->
<div aria-hidden="true" style="position: absolute; left: -5000px;"><input name="b_b2371955307f8826ade8aa973_3d22d9fd4e" tabindex="-1" type="text" value=""/></div>
<div class="clear"><input class="button" id="mc-embedded-subscribe" name="subscribe" type="submit" value="Subscribe"/></div>
</div>
</form>
</div>
<!--End mc_embed_signup--></div>
</aside> <aside class="widget widget_recent_entries" id="recent-posts-2"> <h1 class="widget-title">Recent Posts</h1> <ul>
<li>
<a href="http://basketballintelligence.net/2020/01/02/todays-best-nba-reporting-analysis-1-2-20/">Today’s Best NBA Reporting And Analysis   1/2/20</a>
</li>
<li>
<a href="http://basketballintelligence.net/2020/01/01/todays-best-nba-reporting-analysis-1-1-19/">Today’s Best NBA Reporting And Analysis  1/1/20</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/31/todays-best-nba-reporting-analysis-12-31-19/">Today’s Best NBA Reporting And Analysis  12/31/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/30/todays-best-nba-reporting-analysis-12-30-19/">Today’s Best NBA Reporting And Analysis  12/30/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/29/todays-best-nba-reporting-analysis-12-29-19/">Today’s Best NBA Reporting And Analysis  12/29/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/28/todays-best-nba-reporting-analysis-12-28-19/">Today’s Best NBA Reporting And Analysis  12/28/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/27/todays-best-nba-reporting-analysis-12-27-19/">Today’s Best NBA Reporting And Analysis  12/27/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/26/todays-best-nba-reporting-analysis-12-26-19/">Today’s Best NBA Reporting And Analysis  12/26/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/25/todays-best-nba-reporting-analysis-12-25-19/">Today’s Best NBA Reporting And Analysis  12/25/19</a>
</li>
<li>
<a href="http://basketballintelligence.net/2019/12/24/todays-best-nba-reporting-analysis-12-24-19/">Today’s Best NBA Reporting And Analysis   12/24/19</a>
</li>
</ul>
</aside><aside class="widget widget_text" id="text-7"><h1 class="widget-title">FeedSpot Top 100</h1> <div class="textwidget"><img src="http://blog-cdn.feedspot.com/wp-content/uploads/2016/09/Basketball-transparent_216px__.png"/></div>
</aside></div><!-- #secondary .widget-area -->
</div><!-- #main .site-main -->
<div id="colophon-wrap">
<footer class="site-footer" id="colophon" role="contentinfo">
<div class="site-info">
<a href="http://wordpress.org/" rel="generator" title="A Semantic Personal Publishing Platform">Proudly powered by WordPress</a>
<span class="sep"> | </span>
			Theme: Superhero by <a href="https://wordpress.com/themes/" rel="designer">WordPress.com</a>.		</div><!-- .site-info -->
</footer><!-- #colophon .site-footer -->
</div><!-- #colophon-wrap -->
</div><!-- #page .hfeed .site -->
<script src="http://basketballintelligence.net/wp-content/plugins/simple-share-buttons-adder/js/ssba.js?ver=5.3.2" type="text/javascript"></script>
<script type="text/javascript">
Main.boot( [] );
</script>
<script src="http://basketballintelligence.net/wp-content/themes/superhero/js/small-menu.js?ver=20120206" type="text/javascript"></script>
<script src="http://basketballintelligence.net/wp-includes/js/wp-embed.min.js?ver=5.3.2" type="text/javascript"></script>
</body>
</html>"""
```
